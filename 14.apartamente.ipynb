# pip install scipy

```python
from scipy import stats
```

```python
result = stats.linregress(suprafete, preturi)
result
```

```python
result.slope, result.intercept
```

```python


plt.plot(suprafete, result.slope * suprafete + result.intercept, c="black");
plt.scatter(suprafete, preturi);
plt.xlabel("Suprafete");
plt.ylabel("Preturi");
```

```python
def f(x):
    a = result.slope
    b = result.intercept
    return a * x + b
```

```python
f(60)
```

```python
plt.plot(suprafete, result.slope * suprafete + result.intercept, c="black");
plt.scatter(suprafete, preturi);
plt.scatter(60, f(60));
plt.xlabel("Suprafete");
plt.ylabel("Preturi");
```
