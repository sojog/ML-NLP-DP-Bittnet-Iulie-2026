# Adică găsiți pe w1, w2 si b

```python
def functie_de_activare(numar):
    if numar >= 0:
        return 1
    else:
        return 0

def OR_NEURON(x1, x2):
    w1 = 1 
    w2 = 1
    b = -0.9
    rezultat = w1 * x1 + w2 * x2 + b
    rezultat =  functie_de_activare(rezultat)
    print(f"{x1} or {x2} = ", rezultat)

OR_NEURON(1, 1)
OR_NEURON(1, 0)
OR_NEURON(0, 1)
OR_NEURON(0, 0)
```
