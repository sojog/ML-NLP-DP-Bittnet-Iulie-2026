```python
# 3. Standardizează coloana review la lowercase și fără spații stângă dreapta textului
df["review"] = df["review"].str.lower().str.strip()
df
```

```python
# 4. Definește două liste:
#    - positive_words: ex. great, excellent, amazing, love, perfect, good, fantastic, happy, recommend
#    - negative_words: ex. bad, terrible, worst, broken, poor, disappointed, awful, hate, slow

# Crează o funcție care va mapa review-urile la 2 categorii: pozitive sau negative după regula
#  - dacă pos_hits >= neg_hits → "positive", altfel "negative"
```

```python
positive_words =  "great, excellent, amazing, love, perfect, good, fantastic, happy, recommend".split(", ")
negative_words = "bad, terrible, worst, broken, poor, disappointed, awful, hate, slow".split(", ")
print(positive_words, negative_words)
```

```python
def eticheteaza(text:str):
    cuvinte = text.strip().split()
    pos_hits = sum( 1 for cuv in cuvinte if cuv in positive_words)
    neg_hits = sum( 1 for cuv in cuvinte if cuv in negative_words)
    return "positive" if pos_hits >= neg_hits else "negative"
```

```python
eticheteaza(df.review[0])
```

```python
df["sentiment"] = df["review"].apply(eticheteaza)
df
```
