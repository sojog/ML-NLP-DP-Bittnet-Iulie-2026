```python
df = pd.read_csv("labeled_reviews.csv", index_col=0)
df
```

```python
x = df["review"] # Serie
y = df["sentiment"] # Serie
```

# 2. Vectorizează review-urile cu TfidfVectorizer.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf_vect = TfidfVectorizer()
tfidf_vect
```

```python
tfidf_vect.fit(x)
```

```python
x_tfidf = tfidf_vect.transform(x)
x_tfidf
```

```python
#x_tfidf = tfidf_vect.fit_transform(x)
#x_tfidf
```

```python
x_tfidf.shape
```

```python
x_tfidf.toarray()
```

```python
x_transformed = pd.DataFrame(x_tfidf.toarray(), columns=tfidf_vect.get_feature_names_out())
x_transformed 
```

```python
x_transformed.shape
```
