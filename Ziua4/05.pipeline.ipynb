```python
iris_dataset = load_iris()
iris_dataset.keys()
```

```python
print(iris_dataset.DESCR)
```

```python
x = iris_dataset.data
x.shape
```

```python
y = iris_dataset.target
y
```

```python
x.min(), x.max()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
standard_scaler = StandardScaler()
standard_scaler
```

```python
x_transf = standard_scaler.fit_transform(x)
x_transf
```

```python
x_transf.min(), x_transf.max()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
knn_model = KNeighborsClassifier()
knn_model
```

```python
knn_model.fit(x_transf, y)
```

```python
x[0]
```

```python
knn_model.predict(standard_scaler.transform(x[0:1]))
```

# Versiunea cu Pipeline

```python
from sklearn.pipeline import Pipeline
```

```python
knn_pipeline = Pipeline(steps=[('standard_scaler', StandardScaler()),('model', KNeighborsClassifier())])
knn_pipeline
```

```python
knn_pipeline.fit(x, y)
```

```python
knn_pipeline.steps[0][1]
```

```python
knn_pipeline.predict(x)
```
