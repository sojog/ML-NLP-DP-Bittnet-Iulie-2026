#  2. Fast mathematical operations over arrays

```python
range(10)
```

```python
arr = np.array(range(2, 8))
arr
```

```python
arr = np.arange(2, 8)
arr
```

```python
arr = np.arange(2, 8, dtype=np.int8)
arr
```

```python
arr - 1
```

```python
arr ** 2
```
