```python
new_image = Image.fromarray(np_array)
new_image.save("steag_alb.png")
```

```python
np_array = np.ones((HEIGHT, WIDTH, 3), dtype=np.uint8) * 255
```

```python
RED = 	[0,43,127]
YELLOW = [252,209,22]	
BLUE = [206,17, 38]
```

```python
np_array.shape
```

```python
np_array
```

```python
np_array[ : , 0 : WIDTH // 3, : ] = BLUE
np_array[ : , WIDTH // 3 : 2 * (WIDTH // 3), : ] = YELLOW
np_array[ : ,  2 * (WIDTH // 3) : , : ] = RED
```

```python
new_image = Image.fromarray(np_array)
new_image.save("steag_romania.png")
```
