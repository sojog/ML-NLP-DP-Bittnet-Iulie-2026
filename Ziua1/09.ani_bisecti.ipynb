```python
import numpy as np
```

```python
start_date = 1599
stop_date = 2103
```

# 1. Rezolvare clasică 

```python
ani = list(range(start_date, stop_date + 1))
print(ani)
```

```python
bisecti = []

for i in ani:
    if (i % 400 == 0) or ((i % 4 == 0 ) and (i % 100 != 0)):
        bisecti.append(i)

print(bisecti)
```

```python
bisecti = [ i for i in ani if (i % 400 == 0) or ((i % 4 == 0 ) and (i % 100 != 0))]
print(bisecti)
```

# 2. Rezolvare folosind numpy

```python
ani = np.arange(start_date, stop_date + 1)
```

```python
bisecti = ani [ (ani % 400 == 0) | ((ani % 4 == 0 ) & (ani % 100 != 0)) ]
bisecti
```

```python
divizibil_cu_400 = ani % 400 == 0
divizibil_cu_4 = ani % 4 == 0
nedivizibil_cu_100 = ani % 100 != 0

bisecti = ani [ divizibil_cu_400 | (divizibil_cu_4 & nedivizibil_cu_100) ]
bisecti
```
