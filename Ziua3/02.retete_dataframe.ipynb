# Adaugarea unor elemente noi

```python
retete_noi = {
    "prepTimeMinutes" : [4],	"cookTimeMinutes" : [13], "servings" : [12],	"caloriesPerServing" : [200],	"userId" :[4],	"rating" : [4.9],	"reviewCount" :[123]
}
retete_noi
```

```python
retete_noi = pd.DataFrame(retete_noi)
retete_noi
```

```python
numeric_df = pd.concat([numeric_df, retete_noi], axis=0).sort_values(["prepTimeMinutes", "servings"], ascending=[True, False]).reset_index(drop=True)
numeric_df
```
