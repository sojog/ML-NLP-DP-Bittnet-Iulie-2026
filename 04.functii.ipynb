# 1. Map

```python
lista = [10, 20, 30]

# o_noua_list = [10 ** 2, 20 ** 2, 30 ** 2]

def ridica_la_patrat(x):
    return x ** 2


list(map(ridica_la_patrat, lista))
```

```python
list(map(lambda x:x ** 2, lista))
```

# 2. Filter

```python
lista = [10, 20, 30, 40, 50, 60]

def mai_mici_ca_35(x):
    return x < 35

list(filter(mai_mici_ca_35, lista))
```

```python
list(filter(lambda x: x < 35, lista))
```

# 3. Reduce

```python
from functools import reduce
```

```python
lista = [10, 20, 30, 40, 50, 60]
# lista = [10, 40, 50, 60]

def cel_mai_mic(x, y):
    return x if x<y else y

reduce(cel_mai_mic, lista)
```

```python
reduce(lambda x, y:  x if x<y else y, lista)
```

```python
def suma(x, y):
    return x + y

reduce(suma, lista) 
```

```python
reduce(lambda x, y:  x + y, lista)
```

# Versiune Pythonic - list comprehention

```python
o_lista = [10, 20, 30]
noua_lista = []

for i in o_lista:
    noua_lista.append(i ** 2)

noua_lista
```

```python
o_lista = [10, 20, 30]
noua_lista = [i ** 2 for i in o_lista]
noua_lista
```

```python
lista = [10, 20, 30, 40, 50, 60]
noua_lista = [ i for i in lista if i < 35 ]
noua_lista
```

```python
lista = [10, 20, 30, 40, 50, 60]
noua_lista = [ i for i in lista if i < 35 and i > 20 ]
noua_lista
```
