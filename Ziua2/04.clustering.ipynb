```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans 
```

```python
df = pd.read_csv("datasets/4_categories.csv")
df
```

```python
plt.scatter(df["coord_x"], df["coord_y"])
```

```python
kmeans_model = KMeans(n_clusters=4)
kmeans_model
```

```python
x = df[["coord_x", "coord_y"]]

kmeans_model.fit(x)
```

```python
predictions = kmeans_model.predict(x)
set(predictions)
```

```python
plt.scatter(df["coord_x"], df["coord_y"], c=predictions)
```

```python
plt.scatter(df["coord_x"], df["coord_y"], c=df["category"].map({"a":"purple", "b":"yellow", "c":"green", "d":"blue"}));
```

```python
kmeans_model_3 = KMeans(n_clusters=3)
kmeans_model_3.fit(x)
predictions_3 = kmeans_model_3.predict(x)
plt.scatter(df["coord_x"], df["coord_y"], c=predictions_3)
```
