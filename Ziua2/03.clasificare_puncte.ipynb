```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_model = KNeighborsClassifier()
knn_model
```

```python
x_pt_model = df[["coord_x", "coord_y"]]
y_pt_model = df["category"]
knn_model.fit(x_pt_model, y_pt_model)
```

```python
x = df["coord_x"]
y = df["coord_y"]

plt.scatter(x, y, c=df["category"].map({"a":"red", "b":"green"}))

plt.scatter(225, 275, c="black")
knn_model.predict([[255, 275]])
```

```python
x = df["coord_x"]
y = df["coord_y"]

plt.scatter(x, y, c=df["category"].map({"a":"red", "b":"green"}))

pred_x, pred_y = 400, 200
plt.scatter(pred_x, pred_y, c="black")
knn_model.predict([[pred_x, pred_y]])
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree_model = DecisionTreeClassifier()
tree_model
```

```python
x_pt_model = df[["coord_x", "coord_y"]]
y_pt_model = df["category"]
tree_model.fit(x_pt_model, y_pt_model)
```

# pip install graphviz

```python
from sklearn.tree import export_graphviz
import graphviz
```

```python
tree_graph = export_graphviz(tree_model, feature_names=["coord_x", "coord_y"])
tree_graph
```

```python
type(tree_graph)
```

```python
graphviz.Source(tree_graph)
```
