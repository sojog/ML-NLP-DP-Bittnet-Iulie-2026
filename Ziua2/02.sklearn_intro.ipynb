```python
from  sklearn.linear_model import LinearRegression 
```

```python
model = LinearRegression()
model
```

```python
import numpy as np

# suprafete vs pret
apartamente = np.array([ (30, 54_000), (36, 75_000), (49, 120_000), (53, 175_000),  (69, 193_000), (70, 215_000) , (79, 210_000), (83, 245_000), (84, 284_000), (90, 345_000) ])
x = apartamente[:, 0]
y = apartamente[:, 1]
```

```python
x
```

```python
x_2d = x.reshape(-1, 1)
x_2d
```

```python
model.fit(x_2d, y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
from scipy import stats
result = stats.linregress(x, y)
result.slope, result.intercept
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x, y)
plt.plot(x, model.coef_ * x + model.intercept_)
plt.xlabel("Suprafete")
plt.ylabel("Preturi")
```
